## Data Transformations/Feature Engineering

Create column: StoreType as integer mapped from existing Type column
i.e. have 1, 2, 3 values instead of Type columns values A, B, C

Doing here as much smaller dataset than the one being used for visualisations
which will be a combination of all 3 csv files.


```python
#create column:StoreType as integer (Copilot suggested code)
#left as, is would prefer proper naming convention: dictTypeMapping

type_mapping = {"A": 1, "B": 2, "C": 3} 
dfSales_Stores_DataSet_Work["StoreType"] = dfSales_Stores_DataSet_Work["Type"].map(type_mapping).astype("int64")

#check results get first 50 rows
dfSales_Stores_DataSet_Work.head(50)
```
